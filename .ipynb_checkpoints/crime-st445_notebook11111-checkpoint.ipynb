```python
# Total crimes by crime_type
query1 = """
            SELECT crime_type, count(crime_type) AS total_crimes
            FROM crime_london
            GROUP BY crime_type
            HAVING total_crimes > 0
            ORDER BY total_crimes DESC
        """
```

The following table shows the number and proportion of each crime type in London zone 1 from October 2019 to September 2021.

```python
# Execute query
totalcrimes_by_crime_type = pds.read_sql(query1, con)
# Create percentage column
sum_of_crimes = totalcrimes_by_crime_type['total_crimes'].sum()
totalcrimes_by_crime_type['total_crimes%'] = (totalcrimes_by_crime_type['total_crimes']/sum_of_crimes*100).round(1)
display(totalcrimes_by_crime_type)
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Crimes by major category (%) chart

# Barchart
fig3 = px.bar(totalcrimes_by_crime_type, x ='crime_type', y='total_crimes%',
        labels={
                 'crime_type': 'Crime_type',
                 'total_crimes%': 'Crimes (%)'
                },
         title='Total crimes by crime types (%) in London zone 1, 2019.10-2021.9')
fig3.show()
```

 In order to analyze more clearly, we illustrate the proportion by using bar chart in decreasing order shown in the above figure. We find that the most common cause of crime was "anti-social-behaviour",nearly a quarter, followed by "other-theft", "violent-crime", "theft-from-the-person".
 
 Then we plot another bar chart to show the total number of crimes per month from October 2019 to September 2021 in London zone 1.

```python
table_crime_per_month = pds.read_sql('SELECT distinct year_month, count(year_month) as total_crimes_per_month FROM crime_london GROUP BY year_month', con)

fig4 = px.bar(table_crime_per_month, x ='year_month', y='total_crimes_per_month',
        labels={
                 'year_month': 'YYYY-MM',
                 'total_crimes_per_month': 'Total crimes per month'
                },
         title='The total number of crimes per month in London zone 1, 2019.10-2021.9')
fig4.show()
```

 From October 2019 to April 2020, it remained relatively stable, fluctuating around 15k; while it suddenly dropped in April 2020, we guessed that it was because of the outbreak of Covid-19  in Europe, and it slowly recovered in the following months; however, in 2021 after the UK implemented the lockdown measures in January, the number plummeted again, reaching the lowest point, and then slowly recovered.

```python
# Crimes per month by crime_type
query2 = """
            SELECT crime_type, year_month, count(crime_type) AS total_crimes
            FROM crime_london
            GROUP BY crime_type, year_month
            HAVING total_crimes > 0
            ORDER BY crime_type, year_month
        """
```

The table below shows the number of crimes for each crime type in London zone 1 per month. 

```python
# Execute query
crimes_per_month_by_crime_type = pds.read_sql(query2, con)

# Pivot
crimes_by_month = crimes_per_month_by_crime_type.pivot_table(values='total_crimes',
                                index='year_month',
                                columns=['crime_type'])
display(crimes_by_month)
```

```python
# Line plots
fig5 = px.line(crimes_by_month, x=crimes_by_month.index, y=list(crimes_by_month), title='Major crime categories over time')

fig5.show()
```

In the above line graph, we draw several lines with respect to crime type to show the number of crimes over month. In October 2019 "other-theft" was the highest, followed by "theft-from-the-person", and they both decreased dramatically to the lowest point in April 2020, when the Covid-19 became serious in Europe, and recovered in the following months, however, the number dropped again after the British lockdown in January 2021, and the subsequent recovered. "Violent-crime" followed the similar pattern, and we concluded that Covid-19 had weakened these three crime types. However, "Anti-social-behaviour" had the opposite trend before October 2020,first sharply increased to the peak in October 2019, and fell back in following months. Then its number showed a faster descent after the British lockdown in January 2021.

```python
df_victim_race = df_NY.groupby(['Victim Race']).agg('count').reset_index().rename(columns={'index': 'count'})
df_victim_race = df_victim_race[['Victim Race','count']]
display(df_victim_race)

#åˆ æŽ‰unknown
df_victim_race = df_victim_race.drop(index=(df_victim_race.loc[(df_victim_race['Victim Race']=='UNKNOWN')].index))
#ç®—ç™¾åˆ†æ¯”
sum_of_victim_race = df_victim_race['count'].sum()
df_victim_race['victim_races%'] = (df_victim_race['count']/sum_of_victim_race*100).round(1)
df_victim_race = df_victim_race.sort_values(by='count', ascending=False)

fig11 = px.bar(df_victim_race, x ='Victim Race', y='victim_races%',
        labels={
                 'Victim Race': 'Victim Race',
                 'victim_races%': 'victim_races (%)'
                },
         title='Victim Race % in Manhattan, 2019.10-2021.9')
fig11.show()
```

Compared with the proportion of suspects, the proportion of victims appears to be relatively balanced, although the proportion of black people is still the highest, more than 30%. And the relatively low proportions are WHITE and WHITE HISPANIC, at about 27% and 24% respectively. 
Because of the small difference in the proportions, we can not directly conclude whether race is related to victimization, since the population base of each race is different.
So how to determine whether race has an effect on the victim rate? We can apply the data from https://worldpopulationreview.com/boroughs/manhattan-population to get the following information.
![title](manhattan_race_percentage.jpg)

By such information, we can approximately calculate the proportion of the following four main races in Manhattan, and the remaining races account for a very small proportion, and we ignore them.

WHITE: 58.9%

WHITE HISPANIC: 58.9%-50.7%=8.2%

ASIAN / PACIFIC ISLANDER: 10.3%

BLACK+BLACK HISPANIC: 15.5%

Then we obtain the proportion of each race in Manhattan with respect to these four races.
And for each race, we define the probability of that race's victimization as Degree of dangerous, the formula is that

the proportion of that race among victims(victim_races%) / the proportion of that race in Manhattan(races%)


```python
df_victim_race_potential_rate = df_victim_race[['Victim Race','count']]

count_BLACK_BLACK_HISPANIC = df_victim_race[df_victim_race['Victim Race']=='BLACK'].iloc[0,1]+df_victim_race[df_victim_race['Victim Race']=='BLACK HISPANIC'].iloc[0,1]                       

df_victim_race_potential_rate.loc[7] = ['BLACK / BLACK HISPANIC', count_BLACK_BLACK_HISPANIC]
#df_victim_race_potential_rate.append({'Victim Race':'BLACK / BLACK HISPANIC','count':count_BLACK_BLACK_HISPANIC},ignore_index=True)

df_victim_race_potential_rate = df_victim_race_potential_rate.drop(index=(df_victim_race_potential_rate.loc[(df_victim_race_potential_rate['Victim Race']=='AMERICAN INDIAN/ALASKAN NATIVE')].index))
df_victim_race_potential_rate = df_victim_race_potential_rate.drop(index=(df_victim_race_potential_rate.loc[(df_victim_race_potential_rate['Victim Race']=='BLACK')].index))
df_victim_race_potential_rate = df_victim_race_potential_rate.drop(index=(df_victim_race_potential_rate.loc[(df_victim_race_potential_rate['Victim Race']=='BLACK HISPANIC')].index))

#display(df_victim_race_potential_rate)

sum_of_victim_race2 = df_victim_race_potential_rate['count'].sum()
df_victim_race_potential_rate['victim_races%'] = (df_victim_race_potential_rate['count']/sum_of_victim_race2*100).round(1)
df_victim_race_potential_rate = df_victim_race_potential_rate.sort_values(by='count', ascending=False)
#display(df_victim_race_potential_rate)

df_victim_race_potential_rate['races%'] = [15.5,50.7,8.2,10.3]
#display(df_victim_race_potential_rate)

df_victim_race_potential_rate['degree of dangerous'] = (df_victim_race_potential_rate['victim_races%']/df_victim_race_potential_rate['races%']).round(2)
display(df_victim_race_potential_rate)

fig12 = px.bar(df_victim_race_potential_rate, x ='Victim Race', y='degree of dangerous',hover_data=['victim_races%','count'], color='victim_races%',
        labels={
                 'Victim Race': 'Victim Race',
                 'degree of dangerous': 'degree of dangerous'
                },
         title='Degree of dangerous (= victim_races%/races population percentage%) in Manhattan, 2019.10-2021.9')
fig12.show()
                       
```

We find that the degree of dangerous of WHITE HISPANIC and BLACK+BLACK HISPANIC are much higher than the other two. Hence, we suppose that WHITE HISPANIC and BLACK+BLACK HISPANIC are more likely to be victimized. Although WHITE and ASIAN/PACIFIC ISLANDER both have less degree of dangerous, ASIAN/PACIFIC ISLANDER is more likely to victimized than WHITE.

## 4. Compare Security between London Zone 1 and Manhattan 

Now that we have done our separate analysis of London zone 1 and Manhattan, we continue with what we are interested in, comparing the degree of safety in London zone 1 and Manhattan. Given that the dataset of Manhattan only contains the data of crime which were case-closed, correspondingly we delete the unclosed part of the London zone 1 dataset.

We preprocess the data and then calculate the proportion of each specific type of closed-crime.

```python
#ç»“æ¡ˆçš„çŠ¯ç½ªä¸­å„å…·ä½“ç±»åž‹æ‰€å æ¯”ä¾‹
df_crime_type_finish_london = table_crime_london_finish.groupby(['crime_type']).agg('count').reset_index().rename(columns={'index': 'total_crimes'})
df_crime_type_finish_london = df_crime_type_finish_london[['crime_type','total_crimes']]

sum_of_crimes_finish = df_crime_type_finish_london['total_crimes'].sum()
df_crime_type_finish_london['total_crimes%'] = (df_crime_type_finish_london['total_crimes']/sum_of_crimes_finish*100).round(1)
display(df_crime_type_finish_london)
```

We define the probability of crime as the crime rate(cr), in which we only consider closed-crime, the formula is

cr=tc/p*100000

where tc represents the proportion of total crime, p represents the population.

Since we do not know the population of London zone 1, we use the relevant data of inner London to estimate the population of London zone 1. Inner London has an area of 319 square kilometers and a population of 3.536 million, while London zone 1 has an area of 45 square kilometers. We can estimate the population of London zone 1 as

## 5. Cluster different crime area in New York (Manhattan)

People usually split a map into different regions, and identify high risk regions vs low risk regions. Following this intuition, we proceed to automate this process using KMeans clustering algorithms.

```python
from sklearn.metrics import silhouette_samples
import seaborn as sns
results = []
cluster_range = [2,3,4,5]
for c in cluster_range:
    kmeans = KMeans(init="k-means++",n_clusters = c,n_init=100,random_state=0)
    cluster_label = kmeans.fit_predict(normalized_data)
    silhouette_average = silhouette_score(normalized_data,cluster_label)
    results.append([c,silhouette_average])

results = pd.DataFrame(results,columns=["n_clusters","silhouette_score"])
pivot_km = pd.pivot_table(results,index="n_clusters",values="silhouette_score")

plt.figure()
sns.heatmap(pivot_km,annot=True,linewidth=.5,fmt=".3f",cmap="OrRd_r")
plt.tight_layout()
```

For n_clusters=2, the silhouette_score is maximised, hence we refit the latitude and longitude into 2 clusters.

```python
kmeans_sel = KMeans(init="k-means++",n_clusters = 2,n_init=100,random_state=0).fit(normalized_data)
labels = pd.DataFrame(kmeans_sel.labels_)
clustered_data = cluster_data.assign(Cluster = labels)
```

```python
%matplotlib inline

# libraries for model selection and evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc, confusion_matrix, precision_recall_curve, average_precision_score, accuracy_score

# libraries for model fitting
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

# fix the random seed for reproducibility
np.random.seed(100)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5, random_state=0)
```

```python
classifier_names = ["Logistic Regression",
                    "Nearest Neighbors", 
                    "Linear SVM", 
                    "RBF SVM",
                    "Random Forest",
                    "AdaBoost"]

classifiers = [LogisticRegression(C=1., solver='lbfgs'),
               KNeighborsClassifier(3),
               SVC(kernel="linear", C=0.025, probability=True),
               SVC(gamma=2, C=1, probability=True),
               RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
               AdaBoostClassifier()]

pred_prob = {}
pred = {}

# fit individual classifiers
for name, classifier in zip(classifier_names, classifiers):
    np.random.seed(100)
    classifier.fit(X_train,y_train)
    pred_prob[name] = classifier.predict_proba(X_test)[:,1]
    pred[name] = np.where(pred_prob[name] >= 0.5, 1, 0)
```

```python
def get_iterable(x):
    if isinstance(x, list):
        return x
    else:
        return [x]

def plot_roc(y_test, pred_prob, names, plot_size):
    plt.clf()
    fig = plt.figure(figsize=(plot_size,plot_size))
    ax = fig.add_subplot(1, 1, 1)
    
    names = get_iterable(names)
    for name in names:
        fpr, tpr, _ = roc_curve(y_test, pred_prob[name])
        roc_auc = auc(fpr, tpr)
        ax.plot(fpr, tpr, lw=2.5, label= name + ' (area = %0.3f)' % roc_auc)
        
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.plot([0, 1], [0, 1], color='black', lw=1, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.01])
    plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), fontsize=9)
    plt.show();
    
def plot_pr(y_test, pred_prob, names, plot_size):
    ''' 
    plot pr curve(s) in a single pr space
    
    y_test is the actual response and y_pred is the predicted response
    y_test is an array. y_pred is a dataframe of prediction of methods
    y_pred must have all the columns correponding to the names
    plot_size takes a positive value
    '''
    plt.clf()
    fig = plt.figure(figsize=(plot_size,plot_size))
    ax = fig.add_subplot(1, 1, 1)
    
    names = get_iterable(names)
    for name in names:
        precision, recall, _ = precision_recall_curve(y_test,pred_prob[name])
        avg_precision = average_precision_score(y_test, pred_prob[name])
        ax.plot(recall, precision, lw=2.5, label=name + ' (area = %0.3f)' % avg_precision);
         
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.01])
    plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), fontsize=9)
    plt.show(); 
```

```python
plot_roc(y_test, pred_prob, classifier_names,8)
plot_pr(y_test,pred_prob,classifier_names,8)
```

The Highest AUC achived across all of our model is 0.541, indicating the difficulty to predict severe crime purely using gender and race. Nevertheless, Random Forest achieved the highest AUROC and AUPRC of 0.541 and 0.712 respectively, while Logistic Regression/AdaBoost/Linear SVM comes at close second in terms of both metrics.
